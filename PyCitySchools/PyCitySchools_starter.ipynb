```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
#Store each unique school name into an Index
sch_all = school_data_complete["school_name"].unique()
#Call upon the length of said Index
tot_sch = len(sch_all)
sch_all
```

```python
#Calculate the total number of students
#Print the Length of the column.
tot_stu = len(school_data_complete["Student ID"])
tot_stu
```

```python
#Calculate the total budget
#Calculate the Sum of the column.
tot_bgt = school_data["budget"].sum()
tot_bgt
```

```python
#Calculate the average math score
#Calculate the percentage of students with a passing math score (70 or greater)

#Calculate the Avg of the column.
mmath = school_data_complete["math_score"].mean()

#Run a loop to store all Math Scores above 70
mpass = [score for score in school_data_complete["math_score"] if score >= 70]

#Find percent by dividing lenght of mpass by total students and then multiply by 100
print (f"{round(float(len(mpass)/tot_stu),2)*100}%")
```

```python
#Calculate the average reading score
#Calculate the Mean of the column.

#Calculate the Avg of the column.
mread = school_data_complete["reading_score"].mean()

#Run a loop to store all Reading Scores above 70
rpass = [score for score in school_data_complete["reading_score"] if score >= 70]

#Find percent by dividing lenght of rpass by total students and then multiply by 100
print (f"{round(float(len(rpass)/tot_stu),2)*100}%") 
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)

#Store all students who scored above 70 in Math 
mpass2 = school_data_complete.loc[school_data_complete["math_score"] >= 70]

#Store all students who scored above 70 in Math AND Readding
mrpass = mpass2.loc[mpass2["reading_score"] >= 70]

#Store the Total number of students who passed both
totpass = len(mrpass["Student ID"])

#Find percent by dividing totpass by total students and then multiply by 100
print (f"{round(float(totpass/tot_stu),2)*100}%") 
```

```python
#Creating a dataframe to hold the above results
titles = {
    "Total Schools": tot_sch,
    "Total Students": tot_stu,
    "Total Budget": tot_bgt,
    "Average Math Score": round(mmath,2),
    "Average Reading Score": round(mread,2),
    "% Passing Math": round(float(len(mpass)/tot_stu),2)*100,
    "% Passing Reading": round(float(len(rpass)/tot_stu),2)*100,
    "% Overall Passing": round(float(totpass/tot_stu),2)*100,
    }

d_df = pd.DataFrame([titles])
d_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name
#School Type
#use a Groupby operation to store all individual School Names and Types and store them in a variable.
s_summary = school_data_complete.groupby(["school_name"])
s_type = s_summary["type"].first()


#Total Students
#Store Individual student counts by School name using.count
totstu = s_summary["Student ID"].count()


#Total School Budget
#Per Student Budget
#Store Initial budget counts by School name using.count
totsch_bgt = s_summary["budget"].first()
#Calculate Individual Student Budget by dividing Each Individual School budgets by their respective budget counts
stu_bgt = (totsch_bgt)/(totstu)


#Average Math Score
#Average Reading Score
#Average Math scores are the individual mean of each individual schools Math scores.
indvmathscr = s_summary["math_score"].mean()
#Average Reading scores are the individual mean of each individual schools Reading scores.
indvrdgscr = s_summary["reading_score"].mean()

#Credit: https://www.knytes.com/Cleaning-School-Data-with-Pandas/
#School Math %
smathsummary = school_data_complete[school_data_complete['math_score'] >= 70].groupby(['school_name'])
spassmath = [(i/j)*100 for i,j in zip(smathsummary.math_score.count(),totstu)]


#School Read %
sreadsummary = school_data_complete[school_data_complete['reading_score'] >= 70].groupby(['school_name'])
spassread = [(i/j)*100 for i,j in zip(sreadsummary.reading_score.count(),totstu)]


#Overall Passing %
sovrpass = [(i+j)/2 for i,j in zip(spassmath,spassread)]

#Create a dataframe to hold the above results
s_summary_df = pd.DataFrame({
    "School Type": s_type,
    "Total Students": totstu,
    "Total School Budget": totsch_bgt,
    "Per Student Budget": stu_bgt,
    "Average Math Score": round(indvmathscr,2),
    "Average Reading Score":round(indvrdgscr,2),
    "% Passing Math" : spassmath, 
    "% Passing Reading" :spassread, 
    "% Overall Passing" : sovrpass
})


s_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#First we make bins and aloocate each school accordingly
s_bins = [0, 580, 630, 640, 675]
s_group = ["<$580", "$580-629", "$630-639", "$640-675"]
s_summary_df["Per Student Budget"] = s_summary_df["Per Student Budget"].astype('float')
s_summary_df["Spending (Per Student)"] = pd.cut(s_summary_df["Per Student Budget"], s_bins, labels=s_group)

spend_df = s_summary_df.groupby(["Spending (Per Student)"])              
```

```python
#Next we simply display our bins alongside their respective math and reading statistics
sspend_df = spend_df.mean()

sspend_df[["Average Math Score",
                    "Average Reading Score",
                    "% Passing Math", 
                    "% Passing Reading" , 
                    "% Overall Passing"] 
                       ]
```
